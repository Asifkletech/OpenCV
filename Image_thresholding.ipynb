**Image Thresholding :**

Image thresholding is a technique used to segment the image from grayscale image to black and white


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path="/content/Grayscale_Image.tif"
img=cv2.imread(image_path)
image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(image,cmap="gray")
```

```python
ret,threshold1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(threshold1,cmap="gray")
plt.title("Global thresholding")
plt.axis("off")
```

**In the above image**

**1.The pixels>=127 are set to 255(white)**

**2.The pixels<127 re set to 0(black)**



**Otsu thresholding and automatic thresholding**


**Otsu Thresholding**

Otsu thresholding automatically finds optimal threshold values by minimizing the intra-class variance

we dont need to set the threshold manually,it automatically selects the best threshold

```python
threshold2=ret, threshold2 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(threshold2,cmap="gray")
```

**Adaptive Mean threshold**

The threshold of the each pixel is the mean of the neighbouring pixel

Different threshold values for different regions of the image

```python
threshold3=cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
plt.imshow(threshold3,cmap="gray")
```

**Adaptive Gaussian Thresholding**

Uses the Gaussian wieghted sum of neighbourhood pixels instead of simple mean

Gives better result for the images with shadows and gradient

```python
threshold4=cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
plt.imshow(threshold4,cmap='gray')
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.imshow(threshold1,cmap="gray")
plt.title("Global Thresholding")
plt.axis("off")

plt.subplot(222)
plt.imshow(threshold2,cmap="gray")
plt.title("Outsu Thresholding")
plt.axis("off")

plt.subplot(223)
plt.imshow(threshold3,cmap="gray")
plt.title("Adaptive mean")
plt.axis("off")

plt.subplot(224)
plt.imshow(threshold4,cmap="gray")
plt.title("Adaptive mean")
plt.axis("off")
```
